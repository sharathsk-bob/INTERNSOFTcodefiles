```python
var = pd.DataFrame({'normal': np.random.normal(size=20), 
                       'gamma': np.random.gamma(1, size=20), 
                       'poisson': np.random.poisson(size=20)})
print(var)

var.cumsum(0).plot()
```

```python
var.cumsum(0).plot(subplots=True)
```

```python
var.cumsum(0).plot(secondary_y='gamma')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 12))
for i,vari in enumerate(['normal','gamma','poisson']):
    var[vari].cumsum(0).plot(ax=axes[i], title=vari)
axes[0].set_ylabel('cumulative sum')
```

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='Reds')
plt.colorbar()
```
